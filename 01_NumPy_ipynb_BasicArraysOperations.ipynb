````python
# Subtracting two arrays with different shapes
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([7, 8])  # --> Referenced later

# This will result in an error
# c = a - b

# To make the subtraction possible, we need to broadcast the smaller array
c = a - b[:, np.newaxis]

print(c)
````

````python
# This code snippet demonstrates how to subtract two NumPy arrays with different shapes.

# Create two arrays
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([7, 8])

# Try to subtract the arrays without broadcasting
# This will result in an error
# c = a - b

# To make the subtraction possible, we need to broadcast the smaller array
c = a - b[:, np.newaxis]

# Print the result
print(c)

# Explanation:

# When subtracting two NumPy arrays, they must have the same shape. However, in this case, the shapes of `a` and `b` are different:
# - `a` has a shape of (2, 3)
# - `b` has a shape of (2,)

# To make the subtraction possible, we need to broadcast the smaller array (`b`) to match the shape of the larger array (`a`).
# Broadcasting is a technique that allows us to perform operations on arrays of different shapes by adding missing dimensions to the smaller array.

# In this case, we use the `np.newaxis` object to add a new dimension to `b`. This creates a new array with a shape of (2, 1),
# which can be subtracted from `a`.

# The resulting array `c` has a shape of (2, 3) and contains the element-wise differences between `a` and `b`.
````

## Array Transpose

In machine learning, the numpy.transpose function is often used to manipulate data matrices and tensors in a way that is essential for various operations, such as reshaping data, aligning dimensions for matrix multiplications, and preprocessing data. Practical transpose example:

````python
X = np.array([[1, 2, 3], [4, 5, 6]])
print("Original Array:\n",X)
X_transposed = np.transpose(X)
print("Transposed:\n", X_transposed)
# Result shoud like this
# X_transposed is now:
# array([[1, 4],
#        [2, 5],
#        [3, 6]])
print("Original Array Dimensions:",X.ndim)
print("Transposed Array Dimensions:",X_transposed.ndim)
````

Flattening a 2d array

````python
# Define a 2D array
array_2d = np.array([[1, 2], [3, 4]])

# Flatten the array using np.ravel()
flattened_array = np.ravel(array_2d)

# Display the flattened array and the original array
print("Flattened array:", flattened_array)
print("Original array:\n", array_2d)
````

Lets see a complex example

````python
# Define a matrix X and a vector y
X = np.array([[1, 2, 3], [4, 5, 6]])  # Shape (2, 3)
y = np.array([[7], [8], [9]])  # Shape (3, 1), y is a column vector

# Transpose X to align dimensions for dot product
X_transposed = np.transpose(X)  # Shape (3, 2)

# Perform dot product in the correct order
result = np.dot(y.T, X_transposed)  # y.T has shape (1, 3)

# Display the result
print(result)
````

**Here's how we arrive at [[50 122]]:**

By transposing y (making it a (1, 3) matrix) and multiplying it by X_transposed, the dimensions align correctly for the dot product. This will result in a (1, 2) matrix as the output.

1. Transposing X: We flipped the original X matrix to get X_transposed with shape (3, 2):

[[1, 4],
    [2, 5],
    [3, 6]]

The Dot Product: The dot product is a way of multiplying matrices where you multiply corresponding elements in rows and columns and then sum them up. Since we used y.T (shape (1,3)) and X_transposed (shape (3,2)), the calculation looks like this:

[7  8  9]  *  [[1, 4],
                  [2, 5],
                  [3, 6]]

* To get the first element (50), we do: (7 * 1) + (8 * 2) + (9 * 3) = 50
* To get the second element (122), we do: (7 * 4) + (8 * 5) + (9 * 6) = 122\
That's why the final result is [[50 122]]. It's a 1x2 matrix because of how the dimensions of y.T and X_transposed combined during the dot product.

````python
# @title JSON example
import json

import json

with open('gdrive/My Drive/data/school_data.json') as f:
    for line in f:
        try:
            data = json.loads(line)  # Try loading each line as JSON
            break  # Stop after successfully loading the first object
        except json.JSONDecodeError:
            pass  # Ignore lines that don't contain valid JSON

print(data)  # Print the loaded JSON object
````

1. We iterate through each line in the file.
2. For each line, we attempt to load it as a JSON object using json.loads.
3. If successful, we break out of the loop, assuming we've found the first valid JSON object.
4. If a JSONDecodeError occurs, we simply ignore the line and move on.

This approach should allow you to extract the first valid JSON object from the file, even if there's extra data afterwards.

If you need to handle multiple JSON objects in the file, you can modify the code to store each successfully loaded object in a list or process them as needed.

````python
# Inner Join
inner_join = pd.merge(customers, orders, on='customer_id')
print("Inner Join:")
print(inner_join)
````

## Right Join:
A right join returns all rows from the right table, even if there are no matching rows in the left table.
For unmatched rows in the left table, the corresponding columns will be filled with null values.

````python
# Right Outer Join
right_join = pd.merge(customers, orders, how='right', on='customer_id')
print(right_join)
````

## Left Join:
A left join returns all rows from the left table, even if there are no matching rows in the right table.
For unmatched rows in the right table, the corresponding columns will be filled with null values.

````python
# Left Outer Join
left_join = pd.merge(customers, orders, on='customer_id', how='left')
print("\nLeft Outer Join:")
print(left_join)
````


## Outer Join:
An outer join returns all rows from both the left and right tables, including rows that have no match in the other table.
For unmatched rows, the corresponding columns will be filled with null values.

````python

# Full Outer Join
full_join = pd.merge(customers, orders, on='customer_id', how='outer')
print("\nFull Outer Join:")
print(full_join)
````

````python

# Group By
grouped_data = customers.groupby('city')['customer_name'].count()
print("\nGroup By:")
print(grouped_data)
````
